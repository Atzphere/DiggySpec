```python
import sys
import skimage.io
import numpy as np
import skimage.color
from matplotlib import pyplot as plt
import serial
import time
import csv

#import schedule
timestr = time.strftime("%Y%m%d-%H%M%S")
FILE_NAME = 'Spectumdata'+timestr+'.csv'

f = open(FILE_NAME,'a',newline='')          # open a file for 'a'ppending
writer = csv.writer(f,delimiter=',')        # prepare for writing to file

wavelength_list = ["Wavelength"]

for i in range(312,711):
    wavelength_list.append(i)

writer.writerow(wavelength_list)

hsv = []
value = []

from PIL import Image
import colorsys

for i in range(1,100):
    value = [i]
    im = Image.open(str(i) + ".jpg") # Can be many different formats.
    pix = im.load()
    print(im.size)  # Get the width and hight of the image for iterating over

    sizex = im.size[0]
    sizey = im.size[1]
    
    x = []
    y = []
    r = []
    g = []
    b = []

    for cx in range(sizex):
            cy = sizey/2
            x.append(cx)
            y.append(cx)
            r.append(pix[cx,cy][0])
            g.append(pix[cx,cy][1])
            b.append(pix[cx,cy][2])
    
    for s in range(len(r)):
        value.append(colorsys.rgb_to_hsv(r[s],g[s],b[s])[2])
    writer.writerow(value)
```
